```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.preprocessing import MinMaxScaler
```

Загрузка файлов

```python
df1 = pd.read_excel("../Data/Raw/ebw_data.xlsx")
df1
```

Удаляем идентификатор

Искомые параметры:

    2.Width

Анализ

```python
df1.describe().T
```

```python
import warnings
warnings.filterwarnings("ignore")

#CTGAN,GaussianCopula, CopulaGAN - плохо генерирует
#генерация даных
from sdv.tabular import TVAE
from sdv.evaluation import evaluate


model = TVAE(epochs=1000, 
                 compress_dims=(512,512,512),   
                 decompress_dims=(512,512,512), 
                 embedding_dim=4,
                 l2scale=1e-5, loss_factor=2
                ) 
    
model.fit(df1)
    
```

```python
df_gen_best=model.sample(num_rows=100)
listev = []
listev.append(0)

for x in range(0, 500):
    
    gen_data = model.sample(num_rows=1000)
    ev = evaluate(gen_data, df1)    
    
    if max(listev) < ev:
        df_gen_best = gen_data
    
    listev.append(ev)
    print(ev)

print('вывод')
print(max(listev))
    
```

```python
#объединение датасетов

merged_data = df1.append(df_gen_best)
merged_data
```
